```python
counties_contig = pd.read_csv('../data/usa_counties_contiguous.csv')
counties_contig
```

```python
location = 'http://app02.clerk.org/menu/ccis/Help/CCIS%20Codes/'
filename = 'state_codes.html'
states = pd.read_html(location + filename)[0]

state_code_map = dict(zip(states['Description'], 
                          states['Code']))
counties_contig['StAbbr'] = counties_contig['State'].map(state_code_map)

counties_contig
```

```python
nameDict={"State":"StAbbr"}
counties_sqmi=counties_sqmi.rename(columns=nameDict)
```

```python
counties_sqmi = counties_sqmi[['County', 'StAbbr', 'sq_miles']]
counties_sqmi
```

```python
print(counties_sqmi.columns.values)
print(counties_contig.columns.values)
```

```python
counties_contig['StAbbr'] = counties_contig['StAbbr'].str.strip()
counties_sqmi['StAbbr'] = counties_sqmi['StAbbr'].str.strip()
```

```python
COUNTIES_ALL = pd.merge(counties_sqmi, counties_contig, how = "inner", on = ['County','StAbbr'])
COUNTIES_ALL
```

```python
contig = COUNTIES_ALL.drop_duplicates(subset=['STCOU'], keep='last')
contig
```

```python
contig['Density'] =contig['Population'] / contig['sq_miles']
contig
```

```python
contig.sort_values(by = 'Density')
```

```python
contig.corr()
```

```python
capstone=contig.reindex(columns= ['County', 'State', 'StAbbr', 'STCOU', 'Quality of Life', 'Life Expectancy', 'Median Household Income', 'Percent Some College', ' Percent Exercise', 'Population', 'sq_miles','Density'])
capstone
```

```python
capstone.sort_values(by = 'Quality of Life').head(20)
```

```python
capstone['QOL_rank'] = capstone['Quality of Life'].rank(ascending = 1)
capstone['LE_rank'] = capstone['Life Expectancy'].rank(ascending = 0)
capstone['MHI_rank'] = capstone['Median Household Income'].rank(ascending = 0)
capstone['PSC_rank'] = capstone['Percent Some College'].rank(ascending = 0)
capstone['PE_rank'] = capstone[' Percent Exercise'].rank(ascending = 0)
capstone['SQM_rank'] = capstone['sq_miles'].rank(ascending = 0)
capstone['POP_rank'] = capstone['Population'].rank(ascending = 0)
capstone['DEN_rank'] = capstone['Density'].rank(ascending = 0)
capstone
```

```python
capstone.rename(columns={' Percent Exercise': 'Percent Access to Exercise'}, inplace=True)
capstone.rename(columns={'PE_rank': 'PAE_rank'}, inplace=True)
capstone.rename(columns={'sq_miles': 'Square Miles'}, inplace=True)
capstone
```

```python
capstone.reindex(columns= ['County', 'State', 'StAbbr', 'STCOU', 'Quality of Life', 'Life Expectancy', 'Median Household Income', 'Percent Some College', 'Percent Access to Exercise', 'Square Miles', 'Population', 'Density', 'QOL_rank, LE_rank, MHI_rank', 'PSC_rank', 'PAE_rank', 'SQM_rank'])
pd.set_option('display.max_rows', None)
capstone.head()
```

```python
capstone.to_csv('../data/capstonefull.csv')
```
